```python


import os
import shutil

# Define paths
base_dir = '/content/drive/MyDrive/dataset/GTSRB/parsed_darknet'
output_dir = '/content/drive/MyDrive/dataset/GTSRB/for_yolo'

train_images_dir = os.path.join(output_dir, 'images/train')
test_images_dir = os.path.join(output_dir, 'images/test')
train_labels_dir = os.path.join(output_dir, 'labels/train')
test_labels_dir = os.path.join(output_dir, 'labels/test')

# Create required directories
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(test_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(test_labels_dir, exist_ok=True)

# Function to move files
def move_files(file_type, src_dir, img_dst_dir, label_dst_dir):
    for filename in os.listdir(src_dir):
        if filename.endswith(file_type):
            name, ext = os.path.splitext(filename)
            src_file = os.path.join(src_dir, filename)
            if file_type == '.jpg':
                dst_file = os.path.join(img_dst_dir, filename)
            elif file_type == '.txt':
                dst_file = os.path.join(label_dst_dir, filename)

            shutil.move(src_file, dst_file)

# Move train files
move_files('.jpg', os.path.join(base_dir, 'train'), train_images_dir, train_labels_dir)
move_files('.txt', os.path.join(base_dir, 'train'), train_images_dir, train_labels_dir)

# Move test files
move_files('.jpg', os.path.join(base_dir, 'test'), test_images_dir, test_labels_dir)
move_files('.txt', os.path.join(base_dir, 'test'), test_images_dir, test_labels_dir)

# Create train.txt and test.txt
def create_file_list(images_dir, output_file):
    with open(output_file, 'w') as f:
        for filename in os.listdir(images_dir):
            if filename.endswith('.jpg'):
                file_path = os.path.join(images_dir, filename)
                f.write(file_path + '\n')

create_file_list(train_images_dir, os.path.join(output_dir, 'train.txt'))
create_file_list(test_images_dir, os.path.join(output_dir, 'test.txt'))

print('Dataset organized in Darknet format.')
```
